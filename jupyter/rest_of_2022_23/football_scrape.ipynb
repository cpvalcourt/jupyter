```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

qb_url = 'https://www.cbssports.com/fantasy/football/stats/QB/2022/restofseason/projections/nonppr/'
page = requests.get(qb_url)

soup = BeautifulSoup(page.text, "html.parser")
qb_tab = soup.find("table", {"class":"TableBase-table"})

columns = []
for i in qb_tab.find_all("th", {"class":"TableBase-headTh"}):
    for j in i.find("a"):
        newCol=j.get_text(strip=True)
        if (len(newCol) > 0):
            columns.append(newCol) 
#print(columns)

data = []
for tr in qb_tab.find("tbody").find_all("tr"):
    newRow=[]
    for td in tr.find_all("td")[:1]:
        aText = td.find("a")
        if (aText):
            newRow.append(aText.get_text(strip=True))
    for td in tr.find_all("td")[1:]:
        newRow.append(td.get_text(strip=True))
    data.append(newRow)
df = pd.DataFrame(data, columns=columns)

# print(df)

df.to_csv("qb_data.csv", index=False)
```
